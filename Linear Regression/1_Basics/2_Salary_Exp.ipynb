```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\1_ML\Machine-Learning-\Linear Regression\1_Basics\simple_linear_regression_data.csv")
df.info()
```

```python
# Data visualization

plt.scatter(df.x,df.y,color='black',alpha=0.5)
plt.xlabel("Months ")
plt.ylabel("Expenditure ")
plt.show()
```

```python
# Loss function (dont need in this method )

def loss_function(m,b,points):
    total_error=0
    for i in range (len(points)):
        x= points.iloc[i].x
        y=points.iloc[i].y
        total_error += (y-(m*x+b))**2
    total_error/float(len(points))
```

```python
# Gradient descent Algorithom 

def gradiant(m_now, b_now, points,L):
    m_grad=0
    b_grad=0
    n=len(points)

    for i in range (n):
        x= points.iloc[i].x
        y=points.iloc[i].y

        m_grad += (-2/n)*x *(y-(m_now*x+b_now))
        b_grad += (-2/n) *(y-(m_now*x+b_now))

    m=m_now-m_grad*L
    b=b_now-b_grad*L
    return m,b
```

```python
# Initialization 

m=0
b=0
l=0.0001 
epochs=1000
```

```python
# Training loop 

for i in range (epochs):
    m,b=gradiant(m,b,df,l)

print("the value of optimized m is ",m)
print("the value of optimized b is ",b)
```

```python
# Feeting the Line on graph 

plt.scatter(df.x,df.y,color='black',alpha=0.5)
plt.plot(df.x,(m*df.x+b),color='green')
plt.show()
```
