```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv(r"D:\1_ML\Machine-Learning-\Linear Regression\1_Basics\Experience-Salary.csv")
```

```python
# Equation of stright line is y= wx + b 

w=0
b=0
a= 0.001        # learning rate 
epoch=1000
x=df['salary(in thousands)'].values
y=df['exp(in months)'].values
n=len(df['exp(in months)'])
w_list=[]
b_list=[]
```

```python
# training loop 
for _ in range(epoch):
    # model equation 
    y_predict = w*x + b

    # loss funcion 
    dw = (-2 / n) * np.sum(x * (y - y_predict))
    db = (-2 / n) * np.sum(y - y_predict)

    # update the value 
    w = w - a * dw
    w_list.append(w)
    b = b - a * db
    b_list.append(b)
     
```

```python
print("the w is ",w)
print("the b is ", b )
```

```python
plt.scatter(x,y,color='black',alpha=0.5)
plt.xlabel("Months ")
plt.ylabel("Expenditure ")
plt.plot(x,(w*x+b),color='g')

plt.show()
```

```python
plt.suptitle("graph for W and b ")
plt.subplot(1,2,1)
plt.plot(w_list)
plt.title("Fig: w")
plt.xlabel("epoch")
plt.ylabel("w")

plt.subplot(1,2,2)
plt.plot(b_list)
plt.title("Fig: b")
plt.xlabel("epoch")
plt.ylabel("b")


plt.show()
```
