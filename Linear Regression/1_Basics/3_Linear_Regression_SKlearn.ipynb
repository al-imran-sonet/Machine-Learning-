# Basic Linear Regression using SKlearn 
- One Feature One Target

```python
# basic linear regression from SKlearn 

import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score
```

```python
df =pd.read_csv(r"D:\1_ML\Machine-Learning-\Linear Regression\1_Basics\Experience-Salary.csv")
```

```python
x=df.iloc[:,0:1].values    # iloc= integer-location 
y=df.iloc[:,1].values 
```

```python
print(x.shape)
print(y.shape) 
print(type(x))
print(type(y))
```

```python
df.describe()
```

```python
# Data visualization

plt.scatter(x,y,color='black',alpha=0.5)
plt.xlabel("Months ")
plt.ylabel("Expenditure ")
plt.show()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
print(x_train.shape)
print(x_test.shape) 
```

```python
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
print(regressor.score(x_train,y_train))
print(regressor.score(x_test,y_test))
```

```python
y_pred=regressor.predict(x_test)
m=regressor.coef_
b=regressor.intercept_
print(m)
print(b)
```

```python
# Fitting the line on the data point  

plt.scatter(x,y,color='black',alpha=0.5)
plt.xlabel("Months ")
plt.ylabel("Expenditure ")
plt.plot(x,(m*x+b),color='g')
plt.show()
```

```python
# Accuracy result 

print("mean absolute error is ",mean_absolute_error(y_test,y_pred))
print("mean square  error is ",mean_squared_error(y_test,y_pred))
print("R2 value is ",r2_score(y_test,y_pred))
```

```python
# Prediction vs Actual 

plt.figure(figsize=(15,7))
plt.plot(y_test,color='g')
plt.plot(y_pred,color='r')
plt.show()
```
